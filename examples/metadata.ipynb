```python
from pathlib import Path
from docling_core.types.doc.document import ContentLayer, DoclingDocument, PictureItem
from rich.console import Console
import os

source_path = Path("../test/data/doc/dummy_doc_with_meta.yaml")

doc = DoclingDocument.load_from_yaml(filename=source_path)
# doc._validate_rules()

console = Console()

for item, i in doc.iterate_items(
    included_content_layers={cl for cl in ContentLayer}, with_groups=True, traverse_pictures=True):
    console.rule(f"[bold]{item.self_ref}")
    if item.meta:

        # showcasing adding a new field programmatically
        if item.self_ref == "#/texts/2":
            item.meta.example_custom_field_added_programmaticaly = True

        console.print("item.meta:")
        console.print(item.meta)
        console.print("item.meta.__pydantic_extra__:")
        console.print(item.meta.__pydantic_extra__)
    if isinstance(item, PictureItem):
        console.print(item.annotations)
```
